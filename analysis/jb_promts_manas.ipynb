```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# loading the jailbreak_promts dataset
data = pd.read_csv("../datasets/jailbreak_prompts.csv")
```

```python
# Display the first few rows of the dataset
print(data.head())

# Basic data statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())
```

```python
# Plot the distribution of platforms
plt.figure(figsize=(8, 6))
sns.countplot(data['platform'])
plt.title("Distribution of Platforms")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the distribution of community names
plt.figure(figsize=(12, 6))
sns.countplot(data['community_name'], order=data['community_name'].value_counts().index[:10])
plt.title("Top 10 Community Names")
plt.xlabel("Community Name")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# Grouping prompts by date and counting jailbreaks per day
daily_jailbreak_counts = data[data['jailbreak'] == True].groupby('date').size()
daily_jailbreak_counts.plot(figsize=(12, 6), title="Jailbreak Prompts Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
