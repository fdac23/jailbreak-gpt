```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv('../datasets/questions.csv')
```

```python
# Let's start by getting a summary of the dataset
summary = df.describe()
print("Summary of the dataset:\n", summary)
```

```python
# Calculate the average question length
df['question_length'] = df['question'].apply(lambda x: len(x))
avg_question_length = df['question_length'].mean()
print(f"Average question length: {avg_question_length:.2f} characters")
```

```python
# Create a histogram of question lengths
plt.figure(figsize=(10, 6))
plt.hist(df['question_length'], bins=20, color='skyblue', edgecolor='black')
plt.title("Question Length Distribution")
plt.xlabel("Question Length (Characters)")
plt.ylabel("Frequency")
plt.show()
```
