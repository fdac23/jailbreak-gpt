```python

# Function to count occurrences of words in a file
def count_words(file_path, words):
    with open(file_path, 'r') as file:
        file_contents = file.read().lower()  # Convert to lowercase for case-insensitive matching
    word_counts = {word: file_contents.count(word) for word in words}
    return word_counts

# Replace 'path_to_your_file.txt' with your file's actual path
file_path = 'sample_data/sample.txt'

# Read lines from the file that start with "Message sent:"
with open(file_path, 'r') as file:
    messages = [line.strip() for line in file if line.startswith('Message sent:')]

# Data for the graph
words_to_count = ['alias', 'act', 'pretend', 'character', 'role']
counts = count_words(file_path, words_to_count)

# Sorting the words and occurrences based on occurrences
sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)
words = [x[0] for x in sorted_counts]
occurrences = [x[1] for x in sorted_counts]

# Creating the bar graph
plt.figure(figsize=(8, 5))
bars = plt.bar(words, occurrences, color=['blue', 'green', 'red', 'yellow', 'pink'], edgecolor='black')

# Adding value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center')

# Adding titles and labels
plt.title('Occurrences of Specific Actions in Text Data (Sorted)')
plt.xlabel('Action')
plt.ylabel('Number of Occurrences')

# Rotating x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Display the graph
plt.tight_layout()
plt.show()
```
